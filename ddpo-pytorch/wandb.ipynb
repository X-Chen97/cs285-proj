```python
import wandb
```

```python
api = wandb.Api()
run = api.run("cx9/ddpo-pytorch/b89nfpve")
```

```python
data = run.history()
```

```python
import pandas as pd
```

```python
df_raw = pd.concat(data.dropna(subset='images').apply(lambda x: pd.DataFrame(x['images'])[['filenames', 'captions']], axis=1).to_list()).reset_index()
df_raw
```

```python
# split captions using ' | ', add the prompt and reward columns
caption_df = df_raw['captions'].str.split(' \| ', expand=True).rename(columns={0: 'prompt', 1: 'reward'})
image_df = df_raw['filenames'].str.split('/').str[-1].str.split('.').str[0].str.split('_', expand=True)[[1, 2]].rename(columns={1: 'step', 2: 'name'})
df = pd.merge(image_df, caption_df, left_index=True, right_index=True)
```
